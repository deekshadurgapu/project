```python
import numpy as np 
import pandas as pd
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(2)
```

```python
data = pd.read_csv('E:\\creditcard.csv')
```

```python
data
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
data ['normalizeAmount'] = StandardScaler().fit_transform(data['Amount'].values.reshape(-1,1))
data = data.drop(['Amount'],axis=1)
```

```python
data.head()
```

```python
data = data.drop(['Time'],axis=1)
data.head()
```

```python
x = data.iloc[:, data.columns != 'Class']
y = data.iloc[:, data.columns == 'Class']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.3,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(n_estimators=30)
random_forest.fit(x_train,y_train.values.ravel())
```

```python
y_pred = random_forest.predict(x_test)
```

```python
random_forest.score(x_test,y_test)
```

```python
import pickle
result_rf=random_forest.score(x_test,y_test)

with open('rf.pkl','wb') as my_file_obj:
    pickle.dump(result_rf,my_file_obj)
print('FILE STORED SUCCESSFULLY')
```

```python
cnf_matrix = confusion_matrix(y_test,y_pred)
```

```python
LABELS = ['Normal', 'Fraud']
sns.heatmap(cnf_matrix, xticklabels = LABELS, yticklabels = LABELS, annot = True, fmt =".3f", cmap= "RdPu")
plt.show
```
